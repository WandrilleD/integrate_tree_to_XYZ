```python
import pandas as pd

prefix = 'test6'

dfL = pd.read_csv(prefix+'.leaves.csv',index_col = 0)
dfI = pd.read_csv(prefix+'.internal.csv',index_col = 0)
dfB = pd.read_csv(prefix+'.branches.csv',index_col = 0)
```

```python
x_lines = []
y_lines = []
z_lines = []

for i,row in dfB.iterrows():
    x_lines.extend( [row.x0 , row.x1 , None] )
    y_lines.extend( [row.y0 , row.y1 , None] )
    z_lines.extend( [row.z0 , row.z1 , None] )
```

```python
import plotly.express as px
import plotly.graph_objects as go

trace0 = go.Scatter3d( x = dfL.x , y = dfL.y, z = dfL.z,
                      mode='markers',
                      marker= {'size': 10},
                      hovertemplate='%{text}',
                      text = dfL.index,
                      name='leaves')
trace1 = go.Scatter3d( x = dfI.x , y = dfI.y, z = dfI.z,
                      mode='markers', 
                      marker= {'size': 3},
                      name = 'internal')


trace2 = go.Scatter3d(
    x=x_lines,
    y=y_lines,
    z=z_lines,
    mode='lines',
    name='tree_lines'
)



fig = go.Figure(data=[trace0,trace1, trace2])


fig['layout']['scene']['aspectmode'] = "data"
fig.write_html('tmp5.html')
fig
```
